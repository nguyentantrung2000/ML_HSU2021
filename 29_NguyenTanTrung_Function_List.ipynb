## Bài 4: Viết hàm có 1 tham số n. Hàm xuất ra tất cả các ước số của n. <br> Sử dụng hàm xuất ra tất cả các ước của 30

```python
def uoc(n):
    for i in range(1, n+1):
        if (n % i == 0):
            print(str(i) + '\n')
uoc(30)
```
