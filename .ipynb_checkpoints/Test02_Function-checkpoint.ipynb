## Bài 1: Viết hàm có 2 tham số a và b. Hàm trả về số lớn nhất

```python
def findMax(a,b):
    if(a>b):
        return a
    else:
        return b
findMax(6,9)
```

## Bài 2: Nhập vào 3 số a, b và c bất kỳ. Xuất ra số lớn nhất sử dụng hàm của Bài 1

```python
def findMax(a, b, c):
    max = a
    if max < b: max = b
    if max < c: max = c
    return max
a = int(input('a:'))
b = int(input('b:'))
c = int(input('c:'))
 
print(findMax(a, b, c))


        
```

## Bài 3: Viết hàm có 1 tham số n. Hàm xuất ra bảng cửu chương n.<br> Sử dụng hàm xuất ra bảng cửu chương 9


```python
def cuuChuong(n):
    for i in range(1,11):
        print(n,"x",i,"=",n*i)
cuuChuong(2)
        
    
```
