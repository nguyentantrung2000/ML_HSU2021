```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
# Cau 1A. read data dating site, n_clusters = 3.
dataframe = pd.read_csv('datingTestSet.txt', delimiter = "\t")
print(dataframe.head())
```

```python
# doc du lieu va luot bo bot cot va khoi tao cluster
kmeans = KMeans(n_clusters=3).fit(dataframe.drop(columns=['largeDoses','0.953952'],axis=1))
#fit va luot bo data 2 cot cuoi
# kmeans.fit(dataframe.drop(columns=['largeDoses','0.953952'],axis=1))
```

```python
# cau 1B
x = dataframe['40920']
y = dataframe['8.326976']
fig = plt.figure(figsize=(7,7))
axes = fig.add_subplot(111)

for i in range(x.size):
    if kmeans.labels_[i] == 0: color='red'
    elif kmeans.labels_[i]==1: color='blue'
    else: color='yellow'
    plt.scatter(x[i],y[i],c=color,alpha=1)
    
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='black',marker='^')
plt.scatter([101],[102],c='black')
# tao label cho X va Y va title
plt.xlabel('Fly')
plt.ylabel('Videos Game')
plt.title('KMeans-test')
plt.show()
```

```python
## CÃ¢u 2
## tao list mau cho du lieu
colorList = ['red','blue','yellow']
x_test = [[89,103]]
pred = kmeans.predict(x_test)[0]
#Du doan cluster co dung voi mau cua colorList hay khong ? 
print("Cluster {}: '{}'".format(pred, colorList[pred]))
```
