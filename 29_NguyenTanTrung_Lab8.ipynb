```python
#Train KMeans
kmeans = KMeans(n_clusters=5, random_state=0).fit(X)
print('Centers found by sklearn: ')
print(kmeans.cluster_centers_)
pre_label = kmeans.predict(X)
print(pre_label)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='red')
plt.show()
```

```python
x=X[:,0]
y=X[:,1]
plt.figure(figsize=(7,7), dpi=80)
for i in range(x.size):
    if pre_label[i]==0: color='green'
    elif pre_label[i]==1: color='pink'
    elif pre_label[i]==2: color='blue'
    elif pre_label[i]==3: color='red'
    else: color='yellow'
    plt.scatter(x[i],y[i],c=color,alpha=0.3)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='black',marker='x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('K-means')
plt.show()
```

```python
listColor = ['green','pink','blue','red','yellow']
```

```python
x_test=[[0,100]]
predCluster = kmeans.predict(x_test)[0]
print("Cluster {}:'{}'".format(predCluster,listColor[predCluster]))
```
