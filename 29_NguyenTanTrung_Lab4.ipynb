```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
```

```python
#Câu a
# height (cm)
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183,186,189,192,195,198,201,204]]).T
# weight (kg)
y = np.array([[ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68,69,70,71,72,73,74,75]]).T
```

```python
plt.plot(X,y,'o-r')
plt.show()
```

```python
regr=linear_model.LinearRegression()
regr.fit(X,y) #y= b0 + b1X
b0=regr.intercept_
b1=regr.coef_
print(np.round(b0[0],2))
print(np.round(b1[0][0],2))
print('y = ', np.round(regr.coef_[0][0],3),'x + ', np.round(regr.intercept_[0],3))
```

```python
#Câu b
y177 = regr.predict([[177]])
print(y177)
```

```python
#Câu c 
Y_Predict = regr.predict(X)
plt.plot(X,Y_Predict)
plt.scatter(X,y, color='green')
plt.scatter([[177]],y177, color='red')
```
