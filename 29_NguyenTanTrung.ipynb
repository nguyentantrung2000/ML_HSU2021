## Bài 1: Nhập vào 2 số a và b bất kỳ. Xuất ra số lớn nhất

```python
a = input('Nhap a: ')
b = input('Nhap b: ')
if(a>b):
    print(a)
else:
    print(b)
```

## Bài 2: Nhập vào 3 số a, b và c bất kỳ. Xuất ra số bé nhất

```python
def find_min(a, b, c):
    min = a
    if min > b: min = b
    if min > c: min = c
    return min
 
print("Nhập vào số thứ nhất: ")
a = float(input())
 
print("Nhập vào số thứ hai: ")
b = float(input())
 
print("Nhập vào số thứ ba: ")
c = float(input())
 
print("Số bé nhất trong ba số ", a, " ", b, " ", c, " là ", find_min(a, b, c))
```

## Bài 3: Nhập vào số tự nhiên n. Xuất ra bảng cửu chương n

```python
n = int(input('nhap n: '))
for i in range(1,11):
    print(n,'x',i,'=',n*i)
```

## Bài 4: Nhập vào số tự nhiên n. Xuất ra tất cả các ước số của n

```python
print("Nhập vào số N cần tìm ước: ")
 
n = int(input())
print("Danh sách các ước số của ", n, " là")
for i in range(1, n+1):
    if (n % i == 0):
        print(i, end=",")
```
