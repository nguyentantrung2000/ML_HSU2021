```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([[73.5,75.0,76.0,79.0,81.5,82.5,84.0,85.0,86.5,87.5,89.0,90.0,91.5]]).T

y = np.array([[1.49,1.50,1.51,1.54,1.58,1.59,1.60,1.62,1.63,1.64,1.66,1.67,1.68]]).T

plt.plot(x, y, 'ro')
plt.axis([73.5, 91.5, 1.49, 1.68])
plt.xlabel('Diện tích (m2)')
plt.ylabel('Giá (tỷ đồng)')
plt.title("Scatter")
plt.show()
```
